```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("C:\\Users\\grman\\OneDrive\\Desktop\\BigPapa\\Greg\\BigData\\DataSet\\steam.csv")
fig = plt.figure()
# plt.plot(data["video_60fps"], data["video_240fps"])
# plt.show()
# print(data)

data['genres'] = [i.split(';')[0] for i in data['genres']]

causal_data = data[data['genres'] == 'Casual']
strategy_data = data[data['genres'] == 'Strategy']
racing_data = data[data['genres'] == 'Racing']
rpg_data = data[data['genres'] == 'RPG']

causal_data['release_date'] = [i.split('-')[0] for i in causal_data['release_date']]
strategy_data['release_date'] = [i.split('-')[0] for i in strategy_data['release_date']]
racing_data['release_date'] = [i.split('-')[0] for i in racing_data['release_date']]
rpg_data['release_date'] = [i.split('-')[0] for i in rpg_data['release_date']]

_causal_data = causal_data.groupby('release_date').agg({'price': 'median'}).reset_index()
_strategy_data = strategy_data.groupby('release_date').agg({'price': 'median'}).reset_index()
_racing_data = racing_data.groupby('release_date').agg({'price': 'median'}).reset_index()
_rpg_data = rpg_data.groupby('release_date').agg({'price': 'median'}).reset_index()

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(20, 15))
# Create scatter plot
ax1.plot(_causal_data['release_date'],
         _causal_data['price'],
         mfc='black',
         mec='white',
         mew=2,
         marker='o',
         markersize=12,
         color='crimson')
ax1.set_title('Casual game prices by year')
ax1.grid(linewidth=2, color='mistyrose')
ax2.plot(_strategy_data['release_date'],
         _strategy_data['price'],
         mfc='black',
         mec='white',
         mew=2,
         marker='o',
         markersize=12,
         color='crimson')
ax2.set_title('Strategy games prices by year')
ax2.grid(linewidth=2, color='mistyrose')
ax3.plot(_racing_data['release_date'],
         _racing_data['price'],
         mfc='black',
         mec='white',
         mew=2,
         marker='o',
         markersize=12,
         color='crimson')
ax3.set_title('racing game prices by year')
ax4.grid(linewidth=2, color='mistyrose')
ax4.plot(_rpg_data['release_date'],
         _rpg_data['price'],
         mfc='black',
         mec='white',
         mew=2,
         marker='o',
         markersize=12,
         color='crimson')
ax4.set_title('RPG game prices by year')
ax4.grid(linewidth=2, color='mistyrose')
plt.show()
```
