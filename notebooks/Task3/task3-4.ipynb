```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts 
import numpy as np

data = pd.read_csv("C:\\Users\\grman\\Documents\\BigPapa\\Greg\\BigData\\DataSet\\pr3\\insurance.csv")

bmi = [np.mean(data['bmi']),np.median(data['bmi']),*list(sts.mode(data['bmi']))[0]]
charges = [np.mean(data['charges']),np.median(data['charges']),*list(sts.mode(data['charges']))[0]]

print("bmi anlysis\n")
print("mean = ",bmi[0])
print("median = ",bmi[1])
print("mode = ",bmi[2])

print("charges anlysis\n")
print("mean = ",charges[0])
print("median = ",charges[1])
print("mode = ",charges[2])
```

```python
fig , ax = plt.subplots(1,2, figsize = (15,4))
ax[0].hist(data["bmi"],edgecolor='black')
for i in range(3):
    x,y = [bmi[i],bmi[i]],[330,0]
    ax[0].plot(x,y,linewidth=3)
ax[0].legend(['mean','median','mode'])

ax[1].hist(data["charges"],edgecolor='black')
for i in range(3):
    x,y = [charges[i],charges[i]],[550,0]
    ax[1].plot(x,y,linewidth=3)
ax[1].legend(['mean','median','mode'])
```

```python

bmi_range = [data['bmi'].std(),data['bmi'].max() - data['bmi'].min(),(sts.iqr(data['bmi'],interpolation='midpoint'))]
charges_range = [data['charges'].std(),data['charges'].max() - data['charges'].min(),(sts.iqr(data['charges'],interpolation='midpoint'))]

print("bmi anlysis")
print("Стандартное отклонение = ",bmi_range[0])
print("Размах = ",bmi_range[1])
print("Межквартальный размах = ",bmi_range[2])

print("\nCharges anlysis")
print("Стандартное отклонение = ",charges_range[0])
print("Размах = ",charges_range[1])
print("Межквартальный размах = ",charges_range[2])


```

```python
fig , ax = plt.subplots(1,2, figsize = (15,4))
ax[0].hist(data["bmi"],edgecolor='black')
for i in range(3):
    x,y = [bmi_range[i],bmi_range[i]],[330,0]
    ax[0].plot(x,y,linewidth=3)
ax[0].legend(['ст. откл','размах','меж кв. интервал'])

ax[1].hist(data["charges"],edgecolor='black')
for i in range(3):
    x,y = [charges_range[i],charges_range[i]],[550,0]
    ax[1].plot(x,y,linewidth=3)
ax[1].legend(['ст. откл','размах','меж кв. интервал'])
```
