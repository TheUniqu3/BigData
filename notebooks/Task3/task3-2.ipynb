```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts 
import numpy as np

data = pd.read_csv("C:\\Users\\grman\\Documents\\BigPapa\\Greg\\BigData\\DataSet\\pr3\\insurance.csv")
data.info()
```

```python

# data["children"].hist(color ='blue',edgecolor = 'black')
# data["charges"].hist(color ='blue',edgecolor = 'black')
fig , ax = plt.subplots(1,4, figsize = (15,4))
ax[0].hist(data["age"],edgecolor='black')
ax[0].legend(['age'])
ax[1].hist(data["children"],edgecolor='black')
ax[1].legend(['children'])
ax[2].hist(data["bmi"],edgecolor='black')
ax[2].legend(['bmi'])
ax[3].hist(data["charges"],edgecolor='black')
ax[3].legend(['charges'])
```

```python
bmi = [np.mean(data['bmi']),np.median(data['bmi']),*list(sts.mode(data['bmi']))[0]]
charges = [np.mean(data['charges']),np.median(data['charges']),*list(sts.mode(data['charges']))[0]]
print("bmi anlysis")
print("mean = ",bmi[0])
print("median = ",bmi[1])
print("mode = ",bmi[2])

print("charges anlysis")
print("mean = ",charges[0])
print("median = ",charges[1])
print("mode = ",charges[2])
```

```python
fig , ax = plt.subplots(1,2, figsize = (15,4))
ax[0].hist(bmi,edgecolor='black')
ax[0].legend(['bmi'])
ax[1].hist(charges,edgecolor='black')
ax[1].legend(['charges'])
```
